```python
import tensorflow as tf
import keras 
from keras.models import Sequential 
from keras.layers import Dense,Dropout,Activation
import numpy as np 
```

```python
x_train = np.random.random((1000,20))
y_train = keras.utils.np_utils.to_categorical(np.random.randint(10,size = (1000,1)), num_classes= 10)
x_test = np.random.random((100,20))
y_test = keras.utils.np_utils.to_categorical(np.random.randint(10,size = (100,1)), num_classes= 10)
```

```python
x_train[0]
```

```python
y_train[0]
```

#### Builiding the model 

```python
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=20))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
# compile 
#optimizer = tf.keras.optimizers.SGD(learning_rate = 0.01, decay = lr_decay, momentum = 0.9)
sgt = tf.keras.optimizers.SGD(learning_rate=0.01, decay= 1e-6, momentum=0.9, nesterov = True)
model.compile(loss='categorical_crossentropy',optimizer='sgd', metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,batch_size=100,epochs=1000)
```

```python
model.summary()
```

```python
model.metrics_names
```

```python
score = model.evaluate(x_test, y_test, batch_size= 100)
```

```python
# making prediction 
prediction = model.predict(x_train[0:1])
prediction
```
